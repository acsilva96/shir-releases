```python
import hashlib

def hash_file(file_path):
    # Create a hash object
    hasher = hashlib.sha256()

    # Open the file in binary mode
    with open(file_path, 'rb') as f:
        # Read the file in chunks
        for chunk in iter(lambda: f.read(4096), b''):
            # Update the hash object with the chunk
            hasher.update(chunk)

    # Get the hexadecimal representation of the hash
    file_hash = hasher.hexdigest()

    return file_hash
```

```python
import urllib.request

url = "https://download.microsoft.com/download/E/4/7/E4771905-1079-445B-8BF9-8A1A075D8A10/Release%20Notes.doc"
filename = "/lakehouse/default/Files/SHIR/Release_Notes.doc"

urllib.request.urlretrieve(url, filename)
```

```python
currentFile = '/lakehouse/default/Files/SHIR/currentFile/Release_Notes.doc'
currentFileHash = hash_file(currentFile)

newFileHash = hash_file(filename)

# Compare hashes
if currentFileHash == newFileHash:
    print("All files have the same hash.")
    toRunOrNotToRun = 'noRun'
else:
    print("Files have different hashes.")
    toRunOrNotToRun = 'toRun'
    # need to overwrite old file by new file
print(toRunOrNotToRun)
```

```python
import aspose.words as aw

filename = "/lakehouse/default/Files/SHIR/Release_Notes.doc"
doc = aw.Document(filename)
```

```python
# Save DOC as TXT
converted = "doc-to-text.txt"
doc.save(converted)

#remove "headers" -> remove first 11 rows
with open(converted, 'r',encoding='utf-8') as f:
    lines = f.readlines()


with open(converted, 'w',encoding='utf-8') as f:
    f.writelines(lines[10:])


with open(converted, 'r',encoding='utf-8') as f:
    text = f.read()

sections = text.split('\n\n')

#print(sections)

for i in range(5):
    if i < len(sections) and sections[i]:
        print(sections[i],"\n")
#print(sections[0],"\n")

print("Link to full doc: https://download.microsoft.com/download/E/4/7/E4771905-1079-445B-8BF9-8A1A075D8A10/Release%20Notes.doc \n")
```
